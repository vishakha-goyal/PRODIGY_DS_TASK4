```python
# Load the dataset
df = pd.read_csv("twitter_training.csv")

# Drop rows with missing values
df.dropna(inplace=True)

# Drop irrelevant columns (if any)
# Example: df.drop(columns=['irrelevant_column'], inplace=True)

# Preprocess text data
# Example: Convert text to lowercase, remove special characters, etc.
# df['text'] = df['text'].str.lower()

# Display the cleaned DataFrame
print(df.head())
```

```python
# Visualize sentiment distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Score")
plt.ylabel("Proportion")
plt.show()
```

```python
# Categorize sentiment scores into bins
bins = [-1, -0.5, 0, 0.5, 1]
labels = ['Very Negative', 'Negative', 'Neutral', 'Positive']
df['C'] = pd.cut(df['A'], bins=bins, labels=labels)
```

```python
# Plot distribution of sentiment categories
plt.figure(figsize=(8, 6))
sns.countplot(x='C', data=df, palette="viridis")
plt.title("Distribution of Sentiment Categories")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()
```
